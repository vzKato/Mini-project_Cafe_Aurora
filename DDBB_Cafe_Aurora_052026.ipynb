```python
# Variables de diferentes tipos
# Variables of different types
nombre_negocio = "Café Aurora"        # str (texto)
anio_fundacion = 2019                 # int (entero)
precio_cafe = 45.50                   # float (decimal)
abierto_hoy = True                    # bool (booleano)

# Imprimir información y verificar tipos
# Print information and check types
print("Negocio:", nombre_negocio)
print("Año:", anio_fundacion)
print("Precio del café:", precio_cafe)
print("¿Abierto?:", abierto_hoy)

print("\nTipos de datos:")
print(type(nombre_negocio), type(anio_fundacion), type(precio_cafe), type(abierto_hoy))
```

```python
# Lista de productos que vende la cafetería
# List of products sold by the coffee shop
productos = ["Espresso", "Cappuccino", "Latte", "Americano", "Mocha", "Chai Latte"]

# Lista de ciudades donde hay sucursales
# List of cities with branches
ciudades = ["Querétaro", "CDMX", "Monterrey", "Guadalajara", "Puebla"]

#Diccionario con precios por producto
precios = {
    "Espresso": 35,
    "Cappuccino": 50,
    "Latte": 55,
    "Americano": 40,
    "Mocha": 60,
    "Chai Latte": 58
}

print("Productos:", productos)
print("Ciudades:", ciudades)
print("Precio del Latte:", precios["Latte"])
print("Número de productos:", len(productos))
```

```python
import random
from datetime import datetime, timedelta

# Configuración de semilla para resultados reproducibles
# Seed configuration for reproducible results
random.seed(42)

# Lista vacía donde se guardan las ventas
# Empty list where sales are stored
ventas = []

# Fecha inicial
# Start date
fecha_inicio = datetime(2025, 1, 1)

# Generar 1000 ventas
# Generate 1,000 sales
for i in range (1000):
  fecha = fecha_inicio + timedelta(days=random.randint(0, 180))
  producto = random.choice(productos)
  ciudad = random.choice(ciudades)
  cantidad = random.randint(1, 5)
  precio_unitario = precios[producto]
  total = cantidad * precio_unitario

  venta = {
      "id_venta": i + 1,
      "fecha": fecha.strftime("%Y-%m-%d"),
      "ciudad": ciudad,
      "producto": producto,
      "cantidad": cantidad,
      "precio_unitario": precio_unitario,
      "total": total
  }
  ventas.append(venta)

  # Revisar las primeras 3 ventas
  # Review the first 3 sales
  print("Total de ventas generadas:", len(ventas))
  print("\nPrimeras 3 ventas:")
  for v in ventas[:3]:
    print(v)
```

```python
import pandas as pd
from google.colab import files

# Convertir la lista de diccionarios en un DataFrame
# Convert the list of dictionaries into a DataFrame
df_ventas = pd.DataFrame(ventas)

# Ver las primeras filas
# View the first rows
print(df_ventas.head())
print("\nDimensiones:", df_ventas.shape)

# Guardar como CSV
# Save as CSV
df_ventas.to_csv("ddbb_cafe_aurora.csv")

# Descargar el archivo
# Download the file
files.download("ddbb_cafe_aurora.csv")
```
